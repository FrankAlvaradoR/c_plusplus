```python
#Analizar Texto
%%writefile main.cpp
#include <iostream>
#include <fstream>
using namespace std;
int main(){
  char caracter;
  int cuenta_a=0;
  int cuenta_e=0;
  int cuenta_i=0;
  int cuenta_o=0;
  int cuenta_u=0;
  ifstream archivo("Copia de EjemploTexto.txt",ios::in);
  while(!archivo.eof()){
    archivo.get(caracter);
    if ((caracter=='a') | (caracter=='A')) 
      cuenta_a++;
    if ((caracter=='e') | (caracter=='E')) 
      cuenta_e++;
    if ((caracter=='i') | (caracter=='I')) 
      cuenta_i++; 
    if ((caracter=='o') | (caracter=='O')) 
      cuenta_o++; 
    if ((caracter=='u') | (caracter=='U')) 
      cuenta_u++;      
    cout<<caracter;
  }
archivo.close();
printf("\n \nLa a aparece : %d veces \n",cuenta_a);
printf("La e aparece : %d veces \n",cuenta_e);
printf("La i aparece : %d veces \n",cuenta_i);
printf("La o aparece : %d veces \n",cuenta_o);
printf("La u aparece : %d veces \n",cuenta_u);
}
```

```python
#Ejecuatamos es Script
%%script bash
g++ main.cpp 
./a.out
```
